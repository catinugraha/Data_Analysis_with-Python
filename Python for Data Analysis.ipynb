Now we will try to fill the NaN value for some of the columns.
For the NaN in Num-of-Doors, Bore, and  Stroke columns fill the NaN value with the mean of their columns with the similar value in the Make and Body-Style columns. *e.g if you have NaN in num-of-doors with hatchback renault car then fill the NaN with the num-of-doors mean of the hatchback renault car. 
 

```python
data.fillna(data.groupby(['Make','Body-Style']).transform('mean'), inplace=True)
```

Now, use the IQR method to find the row outlier based on the MPG-Difference columns

```python
Q1_amount = data['MPG-difference'].describe()['25%']
Q3_amount = data['MPG-difference'].describe()['75%']
IQR = Q3_amount - Q1_amount
```

```python
outlier_index = data[(data['MPG-difference'] < Q1_amount - (1.5 * IQR)) | (data['MPG-difference'] >
                                                                           Q3_amount + (1.5 * IQR)) ].index
```

```python
not_outlier_index = data[(data['MPG-difference'] > Q1_amount - (1.5 * IQR)) & (data['MPG-difference'] <
                                                                                 Q3_amount + (1.5 * IQR)) ].index
```

```python
data.loc[outlier_index]
```

```python
sns.distplot(data.loc[outlier_index]['Car-Area'])
plt.title('Car-Area of distribution the outlier')
```

```python
sns.distplot(data.loc[outlier_index]['Price'])
plt.title('Price of distribution the outlier')
```
